```python


from scipy.optimize import root

from src.alg import solve, last

from matplotlib import cm
import matplotlib as mpl
mpl.use('pgf')

from matplotlib.ticker import MultipleLocator

import matplotlib.pyplot as plt
plt.style.use('classic')

pgf_with_rc_fonts = {
    "font.family": "serif",
    "text.usetex": True,
    "pgf.texsystem": "xelatex",
    "pgf.rcfonts": False,
    "pgf.preamble": [
        r"\usepackage{unicode-math}",
        r"\setmainfont{EB Garamond}",
        r"\setmonofont{Courier New}",
        r"\setmathfont{Garamond-Math}",
    ]
}
mpl.rcParams.update(pgf_with_rc_fonts)

import numpy as np
from numpy import linalg as LA
```

```python
MAX_ITER = 100

def newton(z, f, fprime, tol=1e-6):
    for i in range(MAX_ITER):
        step = f(z)/fprime(z)
        if abs(step) < tol:
            return i, z
        z -= step
    return i, z
```

```python
R = 2

def F(z): return z**3 - 1

def dFdz(z): return 3*z**2

def lerp(v0, v1, t):
    return v0 + t / v1

def color(i, z):
    x, y = z.real, z.imag
    # set the "base"-color
    if x > 0: 
        col = 0.0
    else:
        if y > 0:
            col = 0.33
        else:
            col = 0.66
    # map the number of iterations it took to converge
    # to (0, 0.33)
    col += 0.33 * lerp(0, 10, i)
    return col

def color2(i, z):
    x, y = z.real, z.imag
    # set the "base"-color
    if x > 0: 
        col = 0.33
    else:
        if y > 0:
            col = 0.66
        else:
            col = 1.00
    # map the number of iterations it took to converge
    # to (0, 0.33)
    col -= 0.33 * lerp(0, 10, i)
    return col
```

```python
n = 500

M = np.zeros((n,n))

xs = np.linspace(-R, R, n)
ys = np.linspace(-R, R, n)

for i, y in enumerate(xs):
    for j, x in enumerate(ys):
        z = x + y*1j
        M[i, j] = color(*newton(z, F, dFdz))
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
plt.imshow(
    M,
    cmap=cm.viridis,
    vmin = 0.0,
    vmax = 1.0,
    aspect='equal',
    extent=[-R, R, -R, R],
)

fig.savefig("basins.pdf", pad_inches=0, bbox_inches='tight')
```
